```python
df = df.drop(columns=['tweet_id', 'airline_sentiment_confidence', 'negativereason', 'negativereason_confidence', 'airline', 'airline_sentiment_gold', 'name', 'negativereason_gold','retweet_count', 'tweet_coord', 'tweet_created', 'tweet_location', 'user_timezone'])
  
```

```python
df = df[df['airline_sentiment'] != 'neutral']
```

```python
df=df.dropna().reset_index(drop=True)
df
```

```python
df = df.rename(columns = {'text':'review', 'airline_sentiment':'sentiment'})
```

```python
df['sentiment'] = df['sentiment'].map({'positive':1, 'negative': -1 })
df['sentiment'] = df['sentiment'].astype('int')
```

```python
df
```

```python
import re

def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
def cleaning_tweets(tweet):
  clean_tweet = re.sub("@[A-Za-z0-9_]+","", tweet)
  clean_tweet = re.sub("#[A-Za-z0-9_]+","", clean_tweet)

  return clean_tweet
```

```python
df['review'] = df['review'].apply(lambda x: deEmojify(x))
df['review'] = df['review'].apply(lambda x: cleaning_tweets(x))
df
```
